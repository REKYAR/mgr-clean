```python
import pandas as pd
```

```python
df=pd.read_csv("D:/mgr/mgrdeg/project/masters/testing/mgr_tables/mgr_data_v4_all_combined_results.csv")
df
```

```python
filtered_df2 = df.dropna(axis=1, how='any')
```

```python
#drop rows with rho in model_key column
filtered_df2 = filtered_df2[~filtered_df2['model_key'].str.contains('rho', na=False)]
#drop rows that contain without_constraints except if they contain without_constraints_All_Constraints in model_key column
filtered_df2 = filtered_df2[~(filtered_df2['model_key'].str.contains('without_constraints', na=False) & ~filtered_df2['model_key'].str.contains('without_constraints_All_Constraints', na=False))]
```

```python
filtered_df2
```

```python
nonnumeric_columns = filtered_df2.select_dtypes(include=['object']).columns.tolist()
nonnumeric_columns.pop(nonnumeric_columns.index('dataset'))
nonnumeric_columns
```

```python
knn_df = filtered_df2[filtered_df2['directory'].str.startswith('balanced')]
hoeffding_df = filtered_df2[filtered_df2['directory'].str.startswith('hoeffding')]
arf_df = filtered_df2[filtered_df2['directory'].str.contains('arf')]
tree_df = filtered_df2[filtered_df2['directory'].str.contains('sktree')]
#drop results_experiments/ prefix from arf and tree dataframes
arf_df['directory'] = arf_df['directory'].str.replace('results_experiments/', '', regex=False)
tree_df['directory'] = tree_df['directory'].str.replace('results_experiments/', '', regex=False)
```

```python
hoeffding_df_limited = hoeffding_df[['directory','dataset', 'full_name', 'online_accuracy', 'online_f1']]
knn_df_limited = knn_df[['directory','dataset', 'full_name', 'online_accuracy', 'online_f1']]
arf_df_limited = arf_df[['directory','dataset', 'full_name', 'online_accuracy', 'online_f1']]
tree_df_limited = tree_df[['directory','dataset', 'full_name', 'online_accuracy', 'online_f1']]
```

```python
knn_df_limited = knn_df_limited[~knn_df_limited['dataset'].isin(['auto_mpg', 'automobile', 'ionosphere'])]
hoeffding_df_limited = hoeffding_df_limited[~hoeffding_df_limited['dataset'].isin(['auto_mpg', 'automobile', 'ionosphere'])]
arf_df_limited = arf_df_limited[~arf_df_limited['dataset'].isin(['auto_mpg', 'automobile', 'ionosphere'])]
tree_df_limited = tree_df_limited[~tree_df_limited['dataset'].isin(['auto_mpg', 'automobile', 'ionosphere'])]
```

```python
hoeffding_df_average_over_dataset = hoeffding_df_limited[['dataset', 'online_accuracy', 'online_f1']].groupby('dataset').mean().reset_index()
knn_df_average_over_dataset = knn_df_limited[['dataset', 'online_accuracy', 'online_f1']].groupby('dataset').mean().reset_index()
arf_df_average_over_dataset = arf_df_limited[['dataset', 'online_accuracy', 'online_f1']].groupby('dataset').mean().reset_index()
tree_df_average_over_dataset = tree_df_limited[['dataset', 'online_accuracy', 'online_f1']].groupby('dataset').mean().reset_index()
```

```python
hoeffding_df_limited_10k = hoeffding_df_limited[hoeffding_df_limited["directory"].isin(["hoeffding_tree_results_newdata_10000_oh"])]

#for each dataset, pick two rows, one where full name constains "No Constraints" regardless of measures and another, differnt row with the highest accuracy
hoeffding_df_limited_10k_tr_nc = hoeffding_df_limited_10k[hoeffding_df_limited_10k['full_name'].str.contains("No Constraints") ]
#the one with the highest accuracy
hoeffding_df_limited_10k_tr_max = hoeffding_df_limited_10k[hoeffding_df_limited_10k['full_name'].str.contains("No Constraints") == False].groupby('dataset')['online_accuracy'].idxmax()
#union the rows of two max and nc horizontally
hoeffding_df_limited_10k_union = pd.concat([hoeffding_df_limited_10k_tr_nc, hoeffding_df_limited_10k.loc[hoeffding_df_limited_10k_tr_max]], axis=0)
hoeffding_df_limited_10k_union = hoeffding_df_limited_10k_union.reset_index(drop=True)

# move the online_accuracy and online_f1 columns of the No Constrains row to a seprate columns at the max accuracy rows of the same dataset in a new df, remove the No Constraints row
# Separate 'No Constraints' rows and the other rows (which are the max accuracy non-NC rows)
no_constraints_df = hoeffding_df_limited_10k_union[hoeffding_df_limited_10k_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()
#select top accuracy from nc rows for each dataset
no_constraints_df = no_constraints_df.groupby('dataset').first().reset_index()
max_accuracy_non_nc_df = hoeffding_df_limited_10k_union[~hoeffding_df_limited_10k_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()

#rename columns in no_constraints_df to avoid conflicts after merge
no_constraints_df.rename(columns={'online_accuracy': 'online_accuracy_nc', 'online_f1': 'online_f1_nc'}, inplace=True)

#select only relevant columns from no_constraints_df for merging (dataset and the renamed metrics)
no_constraints_to_merge = no_constraints_df[['dataset', 'online_accuracy_nc', 'online_f1_nc']]

#merge the max_accuracy_non_nc_df with the metrics from no_constraints_to_merge
# this adds the 'online_accuracy_nc' and 'online_f1_nc' columns to the max accuracy rows
hoeffding_df_limited_10k_union_max = pd.merge(max_accuracy_non_nc_df, no_constraints_to_merge, on='dataset', how='left')

# join online_accuracy	online_f1 from hoeffding_df_average_over_dataset on dataset as a online_accuracy_averaged and online_f1_averaged
hoeffding_df_limited_10k_union_max = pd.merge(hoeffding_df_limited_10k_union_max, hoeffding_df_average_over_dataset[['dataset', 'online_accuracy', 'online_f1']], on='dataset', how='left')
hoeffding_df_limited_10k_union_max.rename(columns={'online_accuracy_x': 'online_accuracy', 'online_f1_x': 'online_f1', 'online_accuracy_y': 'online_accuracy_averaged', 'online_f1_y': 'online_f1_averaged'}, inplace=True)

hoeffding_df_limited_10k_union_max = hoeffding_df_limited_10k_union_max.drop(columns=['directory'])

hoeffding_df_limited_10k_union_max = hoeffding_df_limited_10k_union_max.round(3)
hoeffding_df_limited_10k_union_max.to_csv("D:/mgr/mgrdeg/project/masters/testing/mgr_tables/mgr_data_ng4_htree.csv", index=False)
```

```python
knn_df_limited_default = knn_df_limited[knn_df_limited["directory"].isin(["balanced_results_multi_newdata_oh_fxs"])]



#for each dataset, pick two rows, one where full name constains "No Constraints" regardless of measures and another, differnt row with the highest accuracy
knn_df_limited_nc = knn_df_limited_default[knn_df_limited_default['full_name'].str.contains("No Constraints") ]
knn_df_limited_nc2 = knn_df_limited_nc.groupby('dataset').first().reset_index()
# the one with the highest accuracy
knn_df_limited_max = knn_df_limited_default[knn_df_limited_default['full_name'].str.contains("No Constraints") == False].groupby('dataset')['online_accuracy'].idxmax()
#union the rows of two max and nc horizontally
knn_df_limited_union = pd.concat([knn_df_limited_nc2, knn_df_limited_default.loc[knn_df_limited_max]], axis=0)
knn_df_limited_union = knn_df_limited_union.reset_index(drop=True)



#separate 'No Constraints' rows from knn_df_limited_union
no_constraints_knn_df = knn_df_limited_union[knn_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()

#ensure one 'No Constraints' row per dataset by taking the first one if multiple exist for a dataset.
no_constraints_knn_df = no_constraints_knn_df.groupby('dataset').first().reset_index()

#separate the other rows (which are the 'max accuracy non-NC' rows) from knn_df_limited_union
max_accuracy_non_nc_knn_df = knn_df_limited_union[~knn_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()

# rename columns in no_constraints_knn_df to avoid conflicts after merge
no_constraints_knn_df.rename(columns={'online_accuracy': 'online_accuracy_nc', 'online_f1': 'online_f1_nc'}, inplace=True)

#select only relevant columns from no_constraints_knn_df for merging
no_constraints_to_merge_knn = no_constraints_knn_df[['dataset', 'online_accuracy_nc', 'online_f1_nc']]

#merge the max_accuracy_non_nc_knn_df with the metrics from no_constraints_to_merge_knn
#this adds 'online_accuracy_nc' and 'online_f1_nc' columns to the max accuracy rows
knn_df_limited_union_max = pd.merge(max_accuracy_non_nc_knn_df, no_constraints_to_merge_knn, on='dataset', how='left')


knn_df_limited_union_max = pd.merge(knn_df_limited_union_max, knn_df_average_over_dataset[['dataset', 'online_accuracy', 'online_f1']], on='dataset', how='left')
knn_df_limited_union_max.rename(columns={'online_accuracy_x': 'online_accuracy', 'online_f1_x': 'online_f1', 'online_accuracy_y': 'online_accuracy_averaged', 'online_f1_y': 'online_f1_averaged'}, inplace=True)
knn_df_limited_union_max = knn_df_limited_union_max.round(3)
knn_df_limited_union_max = knn_df_limited_union_max.drop(columns=['directory'])
knn_df_limited_union_max.to_csv("D:/mgr/mgrdeg/project/masters/testing/mgr_tables/mgr_data_ng4_knn.csv", index=False)
```

```python
# arf
#for each dataset, get rows with "No Constraints" and rows with highest accuracy
arf_df_limited_tr_nc = arf_df_limited[arf_df_limited['full_name'].str.contains("No Constraints")]
arf_df_limited_tr_max = arf_df_limited[~arf_df_limited['full_name'].str.contains("No Constraints")].groupby('dataset')['online_accuracy'].idxmax()

#union the rows
arf_df_limited_union = pd.concat([arf_df_limited_tr_nc, arf_df_limited.loc[arf_df_limited_tr_max]], axis=0)
arf_df_limited_union = arf_df_limited_union.reset_index(drop=True)

# Separate 'No Constraints' rows and the max accuracy rows
arf_no_constraints_df = arf_df_limited_union[arf_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()
arf_no_constraints_df = arf_no_constraints_df.groupby('dataset').first().reset_index()
arf_max_accuracy_df = arf_df_limited_union[~arf_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()

# Rename columns for the merge
arf_no_constraints_df.rename(columns={'online_accuracy': 'online_accuracy_nc', 'online_f1': 'online_f1_nc'}, inplace=True)
arf_no_constraints_to_merge = arf_no_constraints_df[['dataset', 'online_accuracy_nc', 'online_f1_nc']]

# Merge to add NC metrics to the max accuracy rows
arf_df_limited_union_max = pd.merge(arf_max_accuracy_df, arf_no_constraints_to_merge, on='dataset', how='left')

# Add averaged metrics
arf_df_limited_union_max = pd.merge(arf_df_limited_union_max, arf_df_average_over_dataset[['dataset', 'online_accuracy', 'online_f1']], on='dataset', how='left')
arf_df_limited_union_max.rename(columns={'online_accuracy_x': 'online_accuracy', 'online_f1_x': 'online_f1', 'online_accuracy_y': 'online_accuracy_averaged', 'online_f1_y': 'online_f1_averaged'}, inplace=True)

# Clean up and save
arf_df_limited_union_max = arf_df_limited_union_max.drop(columns=['directory'])
arf_df_limited_union_max = arf_df_limited_union_max.round(3)
arf_df_limited_union_max.to_csv("D:/mgr/mgrdeg/project/masters/testing/mgr_tables/mgr_data_ng4_arf.csv", index=False)
```

```python
# SKTree processing
# For each dataset, get rows with "No Constraints" and rows with highest accuracy
tree_df_limited_tr_nc = tree_df_limited[tree_df_limited['full_name'].str.contains("No Constraints")]
tree_df_limited_tr_max = tree_df_limited[~tree_df_limited['full_name'].str.contains("No Constraints")].groupby('dataset')['online_accuracy'].idxmax()

# Union the rows
tree_df_limited_union = pd.concat([tree_df_limited_tr_nc, tree_df_limited.loc[tree_df_limited_tr_max]], axis=0)
tree_df_limited_union = tree_df_limited_union.reset_index(drop=True)

# Separate 'No Constraints' rows and the max accuracy rows
tree_no_constraints_df = tree_df_limited_union[tree_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()
tree_no_constraints_df = tree_no_constraints_df.groupby('dataset').first().reset_index()
tree_max_accuracy_df = tree_df_limited_union[~tree_df_limited_union['full_name'].str.contains("No Constraints", case=False, na=False)].copy()

# Rename columns for the merge
tree_no_constraints_df.rename(columns={'online_accuracy': 'online_accuracy_nc', 'online_f1': 'online_f1_nc'}, inplace=True)
tree_no_constraints_to_merge = tree_no_constraints_df[['dataset', 'online_accuracy_nc', 'online_f1_nc']]

# Merge to add NC metrics to the max accuracy rows
tree_df_limited_union_max = pd.merge(tree_max_accuracy_df, tree_no_constraints_to_merge, on='dataset', how='left')

# Add averaged metrics
tree_df_limited_union_max = pd.merge(tree_df_limited_union_max, tree_df_average_over_dataset[['dataset', 'online_accuracy', 'online_f1']], on='dataset', how='left')
tree_df_limited_union_max.rename(columns={'online_accuracy_x': 'online_accuracy', 'online_f1_x': 'online_f1', 'online_accuracy_y': 'online_accuracy_averaged', 'online_f1_y': 'online_f1_averaged'}, inplace=True)

# Clean up and save
tree_df_limited_union_max = tree_df_limited_union_max.drop(columns=['directory'])
tree_df_limited_union_max = tree_df_limited_union_max.round(3)
tree_df_limited_union_max.to_csv("D:/mgr/mgrdeg/project/masters/testing/mgr_tables/mgr_data_ng4_sktree.csv", index=False)
```
