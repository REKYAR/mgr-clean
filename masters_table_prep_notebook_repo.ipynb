```python
print("start")
path = 'D:/mgr/mgrdeg/project/masters/testing/results_experiments'
h_dirs=['hoeffding_tree_results_newdata_1000',
      'hoeffding_tree_results_newdata_1500',
      'hoeffding_tree_results_newdata_2000',
      'hoeffding_tree_results_newdata_5000',
      'hoeffding_tree_results_newdata_10000']
k_dirs=['balanced_results_multi_newdata', 
        'balanced_results_multi_newdata_aw_40',
        'balanced_results_multi_newdata_cow_40',
        'balanced_results_multi_newdata_constrain_treshold_70']
m_dirs = [
    "arf_basic_newdata",
    "sktree_basic_newdata"
]
m_dirs_oh= [
    "arf_basic_newdata_oh",
    "sktree_basic_newdata_oh"
]

#load files *conbined_results.json in each directory
import json
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns   
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import glob
import os
for h_dir in h_dirs:
    # check if exists
    if not os.path.exists(os.path.join(path, h_dir)):
        print(f"Directory {h_dir} does not exist.")

for k_dir in k_dirs:
    # check if exists
    if not os.path.exists(os.path.join(path, k_dir)):
        print(f"Directory {k_dir} does not exist.")

for m_dir in m_dirs:
    # check if exists
    if not os.path.exists(os.path.join(path, m_dir)):
        print(f"Directory {m_dir} does not exist.")


import re

def standardize_constraint_name(model_key):
    """Standardize constraint naming patterns between KNN and Hoeffding Tree implementations."""
    
    # Handle combined features with positive constraint: Feature1+Feature2(+)
    match = re.search(r'with_constraints_([A-Za-z0-9_\-]+\+[A-Za-z0-9_\-]+)\(\+\)', model_key)
    if match:
        features = match.group(1)
        return f"with_constraints_Positive_Constraint_on_{features}"
    
    # Handle combined features with negative constraint: Feature1+Feature2(-)    
    match = re.search(r'with_constraints_([A-Za-z0-9_\-]+\+[A-Za-z0-9_\-]+)\(\-\)', model_key)
    if match:
        features = match.group(1)
        return f"with_constraints_Negative_Constraint_on_{features}"
    
    # Convert KNN-style single "Feature(-)" to Hoeffding Tree "Negative_Constraint_on_Feature"
    match = re.search(r'with_constraints_([A-Za-z0-9_\-]+)\(\-\)', model_key)
    if match:
        feature = match.group(1)
        return f"with_constraints_Negative_Constraint_on_{feature}"
    
    # Convert KNN-style single "Feature(+)" to Hoeffding Tree "Positive_Constraint_on_Feature"
    match = re.search(r'with_constraints_([A-Za-z0-9_\-]+)\(\+\)', model_key)
    if match:
        feature = match.group(1)
        return f"with_constraints_Positive_Constraint_on_{feature}"
    
    return model_key

def load_and_process_json(file_path, directory_name):
    print(f"Processing file: {file_path}")
    with open(file_path) as f:
        data = json.load(f)
    
    models_data = []
    
    for model_key, model_data in data['models'].items():
        model_name = model_data.get('name', model_key) 
        
        if '_' not in model_key and (model_key == 'with_constraints' or model_key == 'without_constraints'):

            experiment_name = model_name.replace(' ', '_')
            standardized_model_key = f"{model_key}_{experiment_name}"
        else:
            standardized_model_key = model_key
        
        standardized_model_key = standardize_constraint_name(standardized_model_key)
        
        model_record = {
            'directory': directory_name,
            'dataset': data.get('dataset', ''),
            'model_key': standardized_model_key,
            'original_model_key': model_key,  # Keep the original for reference
            'model_name': model_name
        }
        model_record['full_name'] = f"{directory_name}-{model_name}-{data.get('dataset', '')}"
        
        for key, value in model_data.items():
            if key == 'name':
                continue
                
            if isinstance(value, list):
                model_record[key] = value[-1] 
            else:
                model_record[key] = value
                
        models_data.append(model_record)
    
    return pd.DataFrame(models_data)

dfs = {}

# Process all directories
for directory in h_dirs + k_dirs+ m_dirs:
    json_files = glob.glob(os.path.join(path, directory, '*combined_results.json'))
    
    if not json_files:
        print(f"No combined_results.json files found in {directory}")
        continue
        
    for json_file in json_files:
        try:
            df = load_and_process_json(json_file, directory)
            dataset_name = os.path.basename(json_file).split('_combined_results.json')[0]
            if dataset_name:
                key = f"{directory}_{dataset_name}"
            else:
                key = directory
            dfs[key] = df
        except Exception as e:
            print(f"Error processing {json_file}: {e}")
print(f"Processed {len(dfs)} dataframes from directories.")
if dfs:
    all_combined_df = pd.concat(dfs.values(), ignore_index=True)
else:
    all_combined_df = pd.DataFrame()

all_combined_df

all_combined_df.to_csv('mgr_tables/mgr_data_v2_all_combined_results.csv', index=False)
```
